# 01_data_qc

Load index/constituent data, run QC checks, and visualize missingness and spreads.


```python
import sys
from pathlib import Path

ROOT = Path.cwd().resolve()
if (ROOT / "src").exists():
    sys.path.insert(0, str(ROOT))
if (ROOT / "data").exists():
    sys.path.insert(0, str(ROOT))
```

```python
import pandas as pd
import matplotlib.pyplot as plt

from src.io_data import read_market_data

index_csv = "../data/cdx_timeseries.csv"
cons_csv = "../data/constituents_timeseries.csv"

market, bad_dates = read_market_data(index_csv, cons_csv)
print(f"snapshots: {len(market)}")
print(f"bad dates: {bad_dates}")
```

```python
index_df = pd.read_csv(index_csv, parse_dates=["Date"])
cons_df = pd.read_csv(cons_csv, parse_dates=["Date"])

# Normalize column casing for consistency
index_df.rename(columns={"Date": "date", "Tenor": "tenor"}, inplace=True)
cons_df.rename(columns={"Date": "date"}, inplace=True)
index_df["tenor"] = index_df["tenor"].astype(str).str.replace("Y", "", regex=False).astype(float)
index_df["index_spread"] = pd.to_numeric(index_df.get("Index_Mid"), errors="coerce")

missing_by_date = cons_df.isna().groupby(cons_df["date"]).mean().mean(axis=1)
missing_by_date.plot(title="Constituent missing ratio")
plt.ylabel("missing fraction")
plt.show()

# Add a consistent index spread alias for plotting
if "Index_spread" not in index_df.columns:
    index_df["Index_spread"] = pd.to_numeric(index_df.get("Index_Mid"), errors="coerce")
```

```python
subset = index_df[index_df["tenor"] == index_df["tenor"].min()]
subset.plot(x="date", y="index_spread", title="Index spread (shortest tenor)")
plt.show()
```
